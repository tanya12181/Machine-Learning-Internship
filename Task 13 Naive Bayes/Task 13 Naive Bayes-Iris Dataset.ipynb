```python
# Encoding the categorical values with integers
#{'setosa': 1, 'versicolor': 2, 'virginica': 3}
species=np.unique(df['species'])
idx=1
d1,d2={},{}
for s in species:
    d1[s]=idx
    idx+=1
print(d1)
encode={'species':d1}
df=df.replace(encode)
df
```

```python
# df.dtypes
```

```python
# splitting dataset in 80:20 ratio as train and test set respectively
# species is the response variable i.e. column 0
idx = 8*df.shape[0]//10
df = np.array(df)
np.random.shuffle(df)
x_train=df[0:idx,:-1]
y_train=df[0:idx,-1]
x_test=df[idx:,:-1] 
y_test=df[idx:,-1] 
print("\nx train:\n",x_train)
print("\ny train:\n",y_train)
print("\nx test:\n",x_test)
print("\ny test:\n",y_test)
```

```python
from math import sqrt

#finds mean and standard deviation for each column
def find_stats(ds):
    mean=[]
    std=[]
    for c in range(ds.shape[1]):
        mean.append(np.mean(ds[:,c]))
        std.append(np.std(ds[:,c]))
    return mean,std

# We find mean and standard deviation for each class
def stats_by_class(x,y):
    classes = np.unique(y)
    groups={c:[] for c in classes}
    for row in range(y.shape[0]):
        c = y[row]
        groups[c].append(x[row,:])
    
    stats={}
    i=0
    for cl,rows in groups.items():
        mean,std = find_stats(np.array(rows))
        stats[cl]=[mean,std,len(rows)]
        i+=1
    return stats
    
def gaussian_prob(x_i,mean,std):
    return (1/(sqrt(2*np.pi)*std))*np.exp((-1*(x_i-mean)**2)/(2*(std**2)))

def find_prob_for_class(stats,row):
    total_rows=0
    for key in stats.keys():
        total_rows+=stats[key][2]
    prob={}
    for cl in stats.keys():
        prob[cl]=(stats[cl][2]/total_rows)
        #print(prob[cl])
        for i in range(len(stats[cl][0])):
            mean = stats[cl][0][i]
            std = stats[cl][1][i]
            #print(mean,std)
            val = gaussian_prob(row[i],mean,std)
            prob[cl]*=val
    return prob

def predict(stats,test_row):
    prob=find_prob_for_class(stats,test_row)
    #print(prob)
    label,best_prob=None,None
    for key in prob.keys():
        if best_prob is None or best_prob<prob[key]:
            best_prob=prob[key]
            label=key
    return label

def calAccuracy(pred,y_):
    correct=0
    for i in range(y_.shape[0]):
        if pred[i]==y_[i]:
            correct+=1
    return correct*100/y_.shape[0]

def show_predictions(pred,y_):
    for i in range(y_.shape[0]):
        print(f"predicted value: {pred[i]}",end='\t')
        print(f"actual value: {y_[i]}")
```

```python
stats = stats_by_class(x_train,y_train)
predictions=[]
for r in range(x_test.shape[0]):
    label = predict(stats,x_test[r,:])
    predictions.append(label)
accuracy = calAccuracy(predictions, y_test)
print(f"model's accuracy: {accuracy}%")
```

```python
show_predictions(predictions,y_test)
```
