```python
# splitting dataset in 80:20 ratio as train and test set respectively
# species is the response variable i.e. column 0
idx = 8*df.shape[0]//10
df = np.array(df)
np.random.shuffle(df)
x_train=df[:idx,1:]
y_train=df[:idx,0]
x_test=df[idx:,1:] 
y_test=df[idx:,0] 
print("\nx train:\n",x_train)
print("\ny train:\n",y_train)
print("\nx test:\n",x_test)
print("\ny test:\n",y_test)
```

```python
from math import sqrt

#finds mean and standard deviation for each column.Also stores columns 
#with continuous values

def find_stats(x):
    mean,std,columns=[],[],[]
    for col in range(x.shape[1]):
        if isinstance(x[0,col],str)==False:
            mean.append(np.mean(x[:,col]))
            std.append(np.std(x[:,col]))
            columns.append(col)
    return mean,std,columns


#finds log likelihood for category variables
#P(Type|Class)

def find_category_prob(x,y,cls):
    n = x.shape[0]
    probabilities={}
    columns=[]
    probabilities={c:None for c in columns}  
    for col in range(x.shape[1]):
        if isinstance(x[0,col],str):
            types,counts=np.unique(x[:,col],return_counts=True)
            p={t:0 for t in types} #P(Type|Class)
            for r in range(n):
                if y[r]==cls:
                    p[x[r,col]]+=1

            for i in range(types.shape[0]):
                p[types[i]]=np.log(p[types[i]]/counts[i])
            probabilities[col]=p   
        
    return probabilities
    
    
# We find mean and standard deviation for each class
# cls is variable to store the class

def stats_by_class(x,y):
    classes = np.unique(y)
    groups={c:[] for c in classes}
    for row in range(y.shape[0]):
        c = y[row]
        groups[c].append(x[row,:])
    stats={}
    category_prob={}
    for cls,rows in groups.items():
        category_prob[cls]=find_category_prob(x,y,cls)
        mean,std,col = find_stats(np.array(rows))
        stats[cls]=[mean,std,col,len(rows)]
    return stats,category_prob
    
    

    
def gaussian_prob(x_i,mean,std):
    return (1/(sqrt(2*np.pi)*std))*np.exp((-1*(x_i-mean)**2)/(2*(std**2)))



def find_prob_for_class(stats,row,category_prob=None):
    total_rows=0
    for key in stats.keys():
        total_rows+=stats[key][3]
    prob={}
    for cls in stats.keys():
        prob[cls]=np.log(stats[cls][3]/total_rows)
        
        if category_prob is not None:
            d1 = category_prob[cls]
            #print("d1: ",d1)
            for key in d1.keys():
                d2 = d1[key]
                #print("d2: ",d2)
                #print(row[key])
                prob[cls]+=d2[row[key]]
        
        for i in range(len(stats[cls][0])):
            mean = stats[cls][0][i]
            std = stats[cls][1][i]
            idx = stats[cls][2][i]
            #print(mean,std)
            prob[cls]+=np.log(gaussian_prob(row[idx],mean,std))
    return prob


def predict(stats,test_row,category_prob=None):
    posterior=find_prob_for_class(stats,category_prob,test_row)
    #print(prob)
    label,best_posterior=None,None
    for key in posterior.keys():
        if best_posterior is None or best_posterior<posterior[key]:
            best_posterior=posterior[key]
            label=key
    return label



def calAccuracy(pred,y_):
    correct=0
    for i in range(y_.shape[0]):
        if pred[i]==y_[i]:
            correct+=1
    return correct*100/y_.shape[0]



def showPredict(pred,y_):
    for i in range(len(pred)):
        print("predicted: ",pred[i],end="\t")
        print("actual: ",y_[i])
    
```

```python
stats,cat_prob = stats_by_class(x_train,y_train)
predictions=[]
for r in range(x_test.shape[0]):
    label = predict(stats,cat_prob,x_test[r,:])
    predictions.append(label)

accuracy = calAccuracy(predictions, y_test)
print(f"\nmodel's accuracy: {accuracy}%")
```
