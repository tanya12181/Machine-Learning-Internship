```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,11)
y = np.arange(10,21)
a = np.arange(4,8)
b = np.arange(5,9)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('First 2D Graph')
plt.savefig('First Graph.jpeg')
```

```python
y=x*x*x
plt.plot(x,y,'r*',linestyle='-',markersize=10)
```

```python
plt.plot(x,y,'r*',ls='-',markersize=10)
```

```python
plt.plot(x,y,'ro',markersize=10)
```

```python
#creating subplots
b=a*a
plt.subplot(2,2,1)
plt.plot(a,b,'*-')
plt.title('First subplot')

plt.subplot(2,2,2)
plt.plot(a,b,'-o')
plt.title('Second subplot')
```

```python
plt.subplot(2,2,3)
plt.plot(a,b,'*--')
plt.title('Third subplot')

plt.subplot(2,2,4)
plt.plot(a,b,'--o')
plt.title('Fourth subplot')
```

```python
x = np.arange(1,11)
y = 3*x + 5
plt.plot(x,y)
```

```python
x = np.arange(0,np.pi*4,0.1)
y = np.sin(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.title('sin curve')
```

```python
#bar plot
x1 = [3,6,8]
y1 = [5,12,19]

x2 = [2,7,12]
y2 = [4,17,25]

plt.bar(x1,y1,width=0.9,align='edge')
plt.bar(x2,y2,color='g',width=0.9)
```

```python
#bar graph with grid lines
plt.bar(x1,y1,width=1)
plt.bar(x2,y2,color='g',width=1)
plt.grid(b=True,color='black',linestyle='--',which='major',axis='both')
```

```python
plt.barh(x1,y1)
```

```python
#Histograms
a = np.random.randint(10,50,15)
print(a)
plt.hist(a,bins=10)
```

```python
#Pie Chart
labels = ['solution accepted','wrong answer','compile error','partially accepted answer','time limited exceeded']
sizes = [220,550,130,150,50]
colors=['yellowgreen','red','lightgray','yellow','gold']
explode=(0.2, 0,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```

```python
#box plot
data = [np.random.normal(1,std,100) for std in range(1,4)]
#print(data)
plt.boxplot(data,vert=True,patch_artist=True)
```
