```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import seaborn
import urllib.request as ur
```

```
df= pd.read_json('face_detection[1].json',lines=True)
image_links = df['content']
df
```

```
import csv
with open('images_url.csv', 'w',newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["content"])
  for link in image_links:
    writer.writerow([link])
# with open('images.csv','r',newline='') as file:
#   reader = csv.reader(file)
#   for r in reader:
#     print(r)
```

```
#Function for retrieving images from its url and storing it 
#in the specificed file path
#images are numbered (for instance, as image-1.jpg)
def imageFromURL(i,url,file_path):
  file_name=f"image-{i}.jpg"
  full_path=f"{file_path}{file_name}"
  ur.urlretrieve(url,full_path)
```

```
file_name='images_url.csv'
file_path='images/'
urls = pd.read_csv(file_name)
img_no=1
for url in urls['content']:
  if (url!='content'):
    imageFromURL(img_no,url,file_path)
    img_no+=1
```
