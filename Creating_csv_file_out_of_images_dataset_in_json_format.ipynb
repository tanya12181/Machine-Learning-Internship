```
df= pd.read_json('face_detection[1].json',lines=True)
image_links = df['content']
image_links.dtypes
```

```
import csv
with open('images.csv', 'w',newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["content"])
  for link in image_links:
    writer.writerow([link])
with open('images.csv','r',newline='') as file:
  reader = csv.reader(file)
  for r in reader:
    print(r)
```
